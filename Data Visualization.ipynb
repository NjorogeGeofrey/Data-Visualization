```python
#Heatmap
#Generate simple data
data = np.random.rand(10, 6)
rows = list('YXWVUTSRQJ')
columns = list('ABCDEF')

#Basic heatmap
plt.pcolor(data)
plt.show()
```

```python
#Add Row/column Labels
plt.pcolor(data, cmap = plt.cm.Reds, edgecolors='k')
plt.xticks(np.arange(0,6)+0.5,columns)
plt.yticks(np.arange(0,10)+0.5,rows)
plt.show()
```

```python
#Scatter Plot Matrix
from matplotlib.ticker import NullFormatter


df = pd.DataFrame(np.random.randn(1000, 4),
                 columns = ['a', 'b', 'c', 'd'])

spm = pd.plotting.scatter_matrix(df, alpha = 0.2, 
                                      figsize = (10, 7), diagonal = 'hist')
```

```python
# Area Plot
#Generate data

df = pd.DataFrame(np.random.rand(10, 4),
                 columns = ['a', 'b', 'c', 'd'])

plt.figure(figsize =(10, 6))
df.plot(kind = 'area')
```

```python
#Bubble chart
plt.style.use('ggplot')
df = pd.DataFrame(np.random.rand(50, 3), 
                 columns = ['a', 'b', 'c'])

df.plot(kind = 'scatter', x = 'a', y ='b', s = df['c']*400)
```

```python
#3D plot a surface
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
